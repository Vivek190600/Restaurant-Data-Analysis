### 1. Determine the top three most common cuisines in the dataset.

```python
def top_cuisines(df):
    cuisine_counts = df['Cuisines'].value_counts()
    top_3 = cuisine_counts.head(3)
    print("Top 3 Cuisines:", top_3)
top_cuisines(df)
```

```python
# The top 3 Cuisines are 'North Indian', 'North Indian, Chinese' & 'Chinese'
```

### 2. Calculate the percentage of restaurants that serve each of the top cuisines.

```python
def top_cuisines(df):
    print((df['Cuisines'].value_counts().head(3) / len(df)) * 100)
    print("Percentage of Restaurants that serve each of the Top Cuisines:")
top_cuisines(df)
```

```python
# The Percentage of Restaurants that serve each of the Top Cuisines are 'North Indian : 9.8', 'North Indian, Chinese : 5.3' & 'Chinese : 3.7'
```

# Level 2 ---------- Task 2 ---------- Task:  Cuisine Combination

### 1.  Identify the most common combinations of cuisines in the dataset.

```python
def cuisine_combinations(df):
    cuisine_split = df['Cuisines'].str.split(',').explode()
    print("Most Common Cuisine Combinations:")
    print(cuisine_split.value_counts().head(5))
cuisine_combinations(df)
```

### 2. Determine if certain cuisine combinations  tend to have higher ratings.


```python
print("Highest Ratings of Cuisine Combinations:")
cuisine_ratings = df.groupby('Cuisines')['Aggregate_rating'].mean().sort_values(ascending=False)
print(cuisine_ratings.head(5))
```

# Level 2 ---------- Task 3 ---------- Task:  Geographic Analysis

### 1. Plot the locations of restaurants on a map using longitude and latitude coordinates.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Longitude'], y=df['Latitude'], hue=df['City'], palette='coolwarm', alpha=0.6)
plt.title("Restaurant Locations on Map")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

### 2. Identify any patterns or clusters of restaurants in specific areas.

```python
print("Possible Clusters of Restaurants in Specific Areas:")
print(df.groupby(['Latitude', 'Longitude']).size().reset_index(name='Count').sort_values(by='Count', ascending=False))
```
